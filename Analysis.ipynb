### Qst1: What is the best month for sales?

```python
###to make it easy for me i'll make a month column 
```

```python
All_Data['month'] = 3
All_Data.head()
```

```python
All_Data['month'] = All_Data['Order Date'].str[0:2]
All_Data.head()
```

```python
### we have to convert the values of the month columninto integers
```

```python
### but before that we need to remove all NaN and Or values 
```

```python
All_Data = All_Data.dropna(how='all')
All_Data.head()
```

```python
All_Data = All_Data[All_Data['Order Date'].str[0:2] != 'Or']
All_Data.head()
```

```python
### NOW we convert
```

```python
All_Data['month'] = All_Data['Order Date'].str[0:2]
All_Data['month'] = All_Data['month'].astype('int32')
All_Data.head()
```

```python
### in the QST1 we have sales ,so we do exactly as we did with months ,sales = Quantity Ordered*Price each
```

```python
### but first we need to convert their values to numeric so we can do the opertation *
```

```python
All_Data['Quantity Ordered'] = pd.to_numeric(All_Data['Quantity Ordered'])
All_Data['Price Each'] = pd.to_numeric(All_Data['Price Each'])
All_Data.head()
```

```python
### now we create the column sales 
```

```python
All_Data['Sale'] = All_Data['Quantity Ordered'] * All_Data['Price Each']
All_Data.head()
```

```python
### now we can answer qst1
```

```python
Rs = All_Data.groupby('month').sum()
```

```python
import matplotlib.pyplot as plt

month = range(1,13)
plt.bar(month, Rs['Sale'])
plt.xticks(month)
plt.ylabel('Sales')
plt.xlabel('Months')
plt.show()
```

### QST3: what product sold the most?

```python
All_Data.head()
```

```python
product = All_Data.groupby('Product')
```

```python
quantityo = product.sum()['Quantity Ordered']
```

```python
products = [prod for prod, df in product ]

plt.bar( products, quantityo)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product Ordered')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```
